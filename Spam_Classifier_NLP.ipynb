```python
import pandas as pd
```

```python
df = pd.read_csv('SMSSpamCollection',sep='\t',names=['label','sms'])
```

```python
df.head()
```

```python
import nltk
from nltk.stem import WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
import re
```

```python
!python -m nltk.downloader stopwords
```

```python
corpus = []
for i in range(0,len(df)):

  words = re.sub('[^a-zA-Z]','',df['sms'][i])
  words =words.lower()
  words = words.split()
  ps = PorterStemmer()
  words = [ps.stem(word)for word in words if word not in set(stopwords.words('english'))]
  words=''.join(words)
  corpus.append(words)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(max_features=5500)
X = cv.fit_transform(corpus).toarray()
```

```python
y = pd.get_dummies(df['label'])
```

```python
y=y.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
X_train.shape
```

```python
from sklearn.naive_bayes import MultinomialNB
spam_detect=MultinomialNB().fit(X_train,y_train)
```

```python
y_pred = spam_detect.predict(X_test)
```

```python
score = spam_detect.score(X_test,y_test)
score
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
cm = confusion_matrix(y_test,y_pred)
ac = classification_report(y_test,y_pred)
```
